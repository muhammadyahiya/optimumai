## Compute-the-answer exercises (active recall)

```python
from optimumai.exercises.engine import Workbook
wb = Workbook('backprop')
for ex in wb.exercises:
    print(ex.prompt)
    print('  answer:', ex.answer, '-', ex.explanation)
```

```python
# grade your own answer
wb.grade('backprop_1', -3.0).correct
```

## Multiple-choice quiz

```python
from optimumai.quiz.engine import Quiz
q = Quiz('softmax')
for question in q.questions:
    print(question.prompt, '->', question.choices[question.answer])
```

## Make an editable equation ↔ graph (open the HTML in a browser)

```python
from optimumai.visualization.interactive import editable_plot
editable_plot('a*x^2 + b*x + c', out='editable_plot.html')
```
