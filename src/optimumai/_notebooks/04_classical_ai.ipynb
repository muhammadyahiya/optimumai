```python
# The same models as a real fit/predict API:
import numpy as np

from optimumai.ml import KMeans, LinearRegression

X = np.array([[0.0], [1.0], [2.0], [3.0], [4.0]])
y = np.array([1.0, 3.0, 5.0, 7.0, 9.0])          # y = 2x + 1
model = LinearRegression().fit(X, y)
print("prediction for x=10:", float(model.predict(np.array([[10.0]]))[0]))
```

## Visualize it

The explainable traces above are the core. With the `[viz]` extra (`pip install optimumai[viz]`) you can also draw the results — here, k-means clusters and a Sobel edge response computed with `conv2d`.

```python
try:
    import matplotlib.pyplot as plt
    import numpy as np

    from optimumai.ml import KMeans
    from optimumai.vision import conv2d

    fig, axes = plt.subplots(1, 2, figsize=(9, 4))

    rng = np.random.default_rng(0)
    X = np.vstack([
        rng.normal([0, 0], 0.6, (30, 2)),
        rng.normal([3, 3], 0.6, (30, 2)),
    ])
    labels = KMeans(k=2).fit(X).predict(X)
    axes[0].scatter(X[:, 0], X[:, 1], c=labels, cmap="coolwarm", s=18)
    axes[0].set_title("k-means clusters")

    img = np.zeros((16, 16))
    img[4:12, 4:12] = 1.0
    sobel_x = [[1, 0, -1], [2, 0, -2], [1, 0, -1]]
    gx = conv2d(img, sobel_x, padding=1)
    axes[1].imshow(np.abs(gx), cmap="magma")
    axes[1].axis("off")
    axes[1].set_title("vertical-edge response (Sobel-X)")

    plt.tight_layout()
    plt.show()
    print("rendered k-means + Sobel figures")
except ImportError:
    print("For figures:  pip install optimumai[viz]")
```
