## Your turn: fill in the `vector_add` kernel

```python
from optimumai.kernels.exercises import KernelWorkbench

wb = KernelWorkbench()
print(wb.get_challenge('vector_add').prompt)

def my_kernel(ctx, inp, out):
    i = ctx.idx.global_id
    if i < out.size:
        out[i] = ctx.gload(inp['a'], i) + ctx.gload(inp['b'], i)

result = wb.submit('vector_add', my_kernel)
print(result.feedback)
```

Stuck? `wb.reveal('matmul_cell')` shows a correct solution for any challenge.
