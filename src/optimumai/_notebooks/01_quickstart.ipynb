# OptimumAI — Quickstart

Every operation runs with `explain=True` to show the step-by-step math. Start here.

```python
from optimumai import Attention, Matrix, Vector, softmax
```

## Dot product — the atom of matrix multiply

```python
Vector([1, 2, 3]).dot(Vector([4, 5, 6]), explain=True)
```

## Softmax — logits into a probability distribution

```python
softmax([2.0, 1.0, 0.1], explain=True)
```

## Matrix multiplication, cell by cell

```python
Matrix([[1, 2], [3, 4]]).matmul(Matrix([[5, 6], [7, 8]]), explain=True)
```

## Scaled dot-product attention — the transformer core

```python
Attention.demo().render('engineer')
```
